# ANALYSE
De verblijftijd wordt bepaald door de spoeling van het leidingsysteem, vooral bij langdurig niet spoelen ontstaan risico's bijvoorbeeld als iemand op vakantie is.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/Temperatuur_Flow.csv')
df.set_index(pd.to_datetime(df['uitleestijdstip']), inplace=True)
df['volume (l)'] = (df['volume 1'] - df['volume 1'].shift(1))*1000
apps = df['v-app'].unique()

df = df.groupby([pd.Grouper(freq='1D'), 'v-app']).sum().unstack('v-app')
df = df['20190801':]

fig, ax = plt.subplots(nrows=4, figsize=(11, 9))

for idx, app in enumerate(apps[:4]):
    df['volume (l)'][app].plot(ax=ax[idx], ylim=(0,1000), title=app, label='flow (l/dag)', legend=True)
    (df['gelogde min. mediumtemp.'][app]/24).plot(ax=ax[idx], secondary_y=True, label='water temp. (C)', legend=True)
    (df['gelogde gemiddelde omgevingstemp.'][app]/24).plot(ax=ax[idx], secondary_y=True, label='omgevings temp. (C)', legend=True)
    ax[idx].set_ylabel('flow (l/dag)')
    plt.ylabel('temperatuur (C)')
    ax[idx].set_xlabel('datum')
    h1, l1 = ax[idx].get_legend_handles_labels()
    h2, l2 = ax[idx].right_ax.get_legend_handles_labels()
    ax[idx].legend(h1+h2, l1+l2, loc=1, bbox_to_anchor=[1.3,1])
```

```python
df = pd.read_csv('../data/Temperatuur_Flow.csv')
df['uitleestijdstip'] = pd.to_datetime(df['uitleestijdstip'])
df = df.sort_values(['v-app','uitleestijdstip'])
df.set_index('uitleestijdstip', inplace=True)
df['vol'] = df['volume 1']
df['target'] = df['delta (omg-med)']
df['inx'] = df['Unnamed: 0']

current_app = False
current_volume = 0

running_hours = 0
data = []
for inx, row in df.iterrows():
    print(row['inx'], end='\r')
    if row['v-app'] != current_app:
        current_app = row['v-app']
        current_volume = row['vol']
        running_hours = 0
        continue
    if row['vol'] != current_volume:
        current_volume = row['vol']
        running_hours = 0
        continue
    data.append({'hour': running_hours, 'target': row.target})
    running_hours += 1
```

```python
agg = pd.DataFrame(data).groupby('hour').target
agg.median().iloc[:100].plot(label='median delta T (C)', legend=True, figsize=(11,9))
agg.mean().iloc[:100].plot(label='mean delta T (C)', legend=True)
agg.quantile(.75).iloc[:100].plot(label='75 kwantiel (C)', legend=True)
agg.quantile(.25).iloc[:100].plot(label='25 kwantiel (C)', legend=True)
plt.xlabel('datum')
plt.ylabel('delta T (water - omgeving) (C)');
```
