```python
df = pd.read_csv('./../data/Monsteruitslagen_Resultaten.csv')
df['datetime'] = pd.to_datetime(df.datum + ' ' + df.tijd)
df['month'] = df['datetime'].dt.month
df['resultaat'] = pd.to_numeric(df['resultaat'], errors='coerce')
df = df[df.component.str.contains('Totaal Legionella spp')]
df.loc[df['resultaat']>10000, 'resultaat'] = 10000
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(11,9))
sns.boxplot(x='month', y='resultaat', data=df)
```
